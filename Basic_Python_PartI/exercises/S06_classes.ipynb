# Python Classes and Objects
Python is an object oriented programming language. Almost everything in Python is an object, with its properties and methods.

## Class
A Class is like an object constructor, or a "blueprint" for creating objects.

````python
# Create a class named MyClass with a class attribute x set to 5

class MyClass:

    x = 5
````

````python
MyClass
````

## Instance
Instance of a class is an object that is created using the class blueprint.

````python
# Create an instance of MyClass and print the value of x.

p1 = MyClass()
print(p1.x)

p2 = MyClass()
print(p2.x)

print("-"*50)

p2.x = 10

print(p1.x)
print(p2.x)
````

## Constructor
- All classes have a function called <code>\__init__()</code>, which is always executed when the class is being initiated.
- Use the <code>\__init__()</code> function to assign values to object properties, or other operations that are necessary to do when the object is being created:

````python
# Create a class named Person with a constructor that takes name and age as parameters and assigns them to instance attributes.

class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age
````

````python
# Create an instance of Person and print the name and age attributes.

p1 = Person(name="Tom", age=30)
print(f"Name: {p1.name}, Age: {p1.age}")

p2 = Person(name="Marry", age=25)
print(f"Name: {p2.name}, Age: {p2.age}")
````

## Object methods
Objects can also contain methods. Methods in objects are functions that belong to the object.  Let us create a method in the Person class:

````python
# Add a method named myfunc to the Person class that prints a greeting with the person's name.

class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def myfunc(self):
        print(f"Hello, my name is {self.name}.")
````

````python
# Create an instance of Person and call the myfunc method.

p1 = Person(name="Tom", age=30)
p1.myfunc()
print("-"*50)
p2 = Person(name="Zara", age=25)
p2.myfunc()
````

## Example

````python
class Car:
    def __init__(self,brand,horsePower,color):
        self.brand = brand
        self.horsePower = horsePower
        self.color = color
        self.currentPosition = 0
 
    def start(self):
        print("start the car")
    def moveForward(self):
        self.currentPosition = self.currentPosition + self.horsePower
    def moveBackward(self):
        self.currentPosition = self.currentPosition - self.horsePower
````

````python
# Create two instances of the Car class with different attributes.

car_ricky = Car("Honda", 200, "Red")
car_tao = Car("Toyota", 300, "Blue")
````

````python
# Use the methods of car_ricky and car_tao to demonstrate their functionality.

print(car_ricky.currentPosition)
car_ricky.start()
car_ricky.moveForward()
print(car_ricky.currentPosition)
````
