# Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Single Plot


```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(start=0, stop=3 * np.pi, step=0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x, y, "--r*")
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(start=0, stop=3 * np.pi, step=0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the sine and cosine curves
plt.plot(x, y_sin)
# plt.show()
plt.plot(x, y_cos)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["Sine", "Cos"])
plt.title("Sine and Cos")
plt.show()
```

```python
# Alternative way to create the same plot using object-oriented interface

fig, ax = plt.subplots()
ax.plot(x, y_sin, label="Sine")
ax.plot(x, y_cos, label="Cos")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()
```

### Multiple Plots


```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create a figure with 2x2 subplots.
# Contrained layout automatically adjusts subplot parameters to give specified padding

fig, axes = plt.subplots(2, 2, constrained_layout=True, sharex=True, sharey=True)

# Make the first plot
axes[0, 0].plot(x, y_sin)
axes[0, 0].set_title("Sine")

# Make the second plot
axes[1, 1].plot(x, y_cos)
axes[1, 1].set_title("Cos")

axes[0, 1].plot(x, 2 * y_sin, "r")
axes[0, 1].set_title("twice of sine")
axes[1, 0].plot(x, 2 * y_cos, "r")
axes[1, 0].set_title("twice of cos")

# Create title for the whole figure
fig.suptitle("Plot")
```

```python
# Create data for multiple plots
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_sum = y_sin + y_cos  

# Create a figure with 3 rows and 1 column of subplots

fig, axes = plt.subplots(3, 1, constrained_layout=True, sharex=True, sharey=True)

# Make the first plot

axes[0].plot(x, y_sin, linestyle="--", color="red")
axes[0].set_title("Sine")
# Make the second plot

axes[1].plot(x, y_cos, linestyle=":", color="green")
axes[1].set_title("Cos")
# Make the third plot    

axes[2].plot(x, y_sum, linestyle="-.", color="blue")
axes[2].set_title("Sum")

fig.suptitle("Plot")
```
