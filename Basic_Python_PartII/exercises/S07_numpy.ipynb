# Numpy


```python
# Import the numpy library as np

import numpy as np
```

```python
# Create a numpy array with the elements [1, 2, 3, 4, 5]

arr = np.array([1, 2, 3, 4, 5])

print(arr.mean())
print(arr.max())
print(arr.min())
```

```python
# Create a numpy array with the elements 1 to 100.

hundreds_arr = np.arange(1, 100+1, 1)
hundreds_arr
```

```python
# Print the shape of the array from the previous step

hundreds_arr.shape
```

```python
np.ndim(hundreds_arr)
```

```python
# Create a 2x2 numpy array with elements
# |1, 2|
# |3, 4|

arr_2x2 = np.array([[1, 2], [3, 4]])
print(arr_2x2)
print("-"*50)
print(arr_2x2.shape)
print("-"*50)
print(np.ndim(arr_2x2))
```

```python
# Calculate the sum of all elements in the 2x2 array

print(np.sum(arr_2x2))
print("-"*50)
print(arr_2x2.sum(axis=0))
print("-"*50)
print(arr_2x2.sum(axis=1))
```

```python
# Calculate the mean of the elements in the 2x2 array
print(arr_2x2)
print("-"*50)
print(arr_2x2.mean(axis=0))
print("-"*50)
print(arr_2x2.mean(axis=1))
```

```python
# Reshape the 2x2 array to a 1x4 array and print it

print(arr_2x2.reshape(1, 4))
print("-"*50)
print(arr_2x2.reshape(1, 4).shape)
print("-"*50)
print(np.ndim(arr_2x2.reshape(1, 4)))
```

```python
print(arr_2x2.reshape(4, 1))
print("-"*50)
print(arr_2x2.reshape(4, 1).shape)
print("-"*50)
print(np.ndim(arr_2x2.reshape(4, 1)))
```

```python
arr2 = np.arange(1, 5, 1)
print(arr2)
print(arr2.shape)
print(np.ndim(arr2))

print("-"*50)

arr3 = arr2.reshape(2, 2)
print(arr3)
print(arr3.shape)
print(np.ndim(arr3))

print("-"*50)

arr4 = arr3.reshape(4, 1)
print(arr4)
print(arr4.shape)
print(np.ndim(arr4))
```

```python
my_list = []

for row in arr4:
    for elm in row:
        my_list.append(elm)

arr5 = np.array(my_list)

print(arr5)
print(arr5.shape)
print(np.ndim(arr5))
```
