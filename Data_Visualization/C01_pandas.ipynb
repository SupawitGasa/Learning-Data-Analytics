# Pandas


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("data/iris.csv")
```

```python
iris.head()
```

## Basic Data Inspection


```python
iris["species"].value_counts()
```

```python
# Bar plot of species counts
# iris["species"].value_counts().plot.bar()

# Add more customizations
iris['species'].value_counts().plot.bar(
    figsize=(8, 6), color='coral', edgecolor='black'
)
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Alternative syntax using Axes object
ax = iris['species'].value_counts().plot.bar(
    figsize=(8, 6), color='coral', edgecolor='black'
)
ax.set_title('Count of Each Species')
ax.set_xlabel('Species')
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Horizontal bar plot
iris["species"].value_counts().plot.barh()
plt.show()

# Add more customizations to horizontal bar plot
iris['species'].value_counts().plot.barh(
    figsize=(8, 6), color='lightgreen', edgecolor='black'
)
plt.title('Count of Each Species (Horizontal)')
plt.xlabel('Count')
plt.ylabel('Species')
plt.tight_layout()
plt.show()
```

```python
# Pie chart of species counts
iris["species"].value_counts().plot.pie()
plt.show()

# Add more customizations to pie chart
iris["species"].value_counts().plot.pie(
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['lightblue', 'lightgreen', 'lightcoral']
)
plt.show()
```

```python
# Plot histogram for a single column
iris["sepal_length"].hist()
plt.show()

# Add more customizations to histogram
iris['sepal_length'].hist(bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for all numeric columns
iris.hist()
plt.show()
# Try choosing specific columns
iris[['sepal_length', 'sepal_width']].hist()
plt.show()

# Add more customizations to multiple histograms
iris.hist(bins=15, edgecolor='black', figsize=(10, 5))
plt.show()
```

```python
# Box plot for all numeric features
iris.boxplot()
plt.show()

# Add more customizations to box plot
iris.boxplot(figsize=(10, 6), grid=False)
plt.title('Box Plot of Iris Features')
plt.ylabel('Value (cm)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Box plot by species
iris.boxplot(by="species")
plt.show()

# Add more customizations to box plot by species
iris.boxplot(by='species', figsize=(12, 8), layout=(2, 2), grid=False)
plt.suptitle('Box Plots by Species', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot with color by species
iris.plot.scatter(x="sepal_length", y="sepal_width")
plt.show()

# Add more customizations to scatter plot
colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}
iris.plot.scatter(
    x='sepal_length', 
    y='sepal_width', 
    c=iris['species'].map(colors), 
    figsize=(8, 6)
)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Scatter matrix to see all pairwise relationships
from pandas.plotting import scatter_matrix

scatter_matrix(iris)
plt.show()

# Add more customizations to scatter matrix
scatter_matrix(
    iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]],
    figsize=(12, 12),
    diagonal="hist",
    alpha=0.5,
)
plt.suptitle("Scatter Matrix of Iris Features", y=0.995, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Density plot for all numeric columns
iris.plot.kde()
plt.show()

# Add more customizations to density plot
iris.plot.kde(figsize=(10, 6), alpha=0.7)
plt.title("Density Plot of Iris Features")
plt.xlabel("Value (cm)")
plt.ylabel("Density")
plt.legend(loc="best")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
