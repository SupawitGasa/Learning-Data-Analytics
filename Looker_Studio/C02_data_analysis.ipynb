```python
import pandas as pd
```

```python
# sales = pd.read_excel("sales_customer.xlsx")
sales = pd.read_excel("sales_customer_small.xlsx")
```

```python
# Display the first few rows of the DataFrame
sales.head()
```

```python
# Get the shape of the DataFrame
sales.shape
```

```python
# Get the column names of the DataFrame
sales.columns
```

```python
# Get summary statistics of the DataFrame
sales.describe()
```

```python
# Count occurrences of each product category
sales["Product Category"].value_counts()
```

```python
# Inspect the number of records for each combination of Product Category, Sub Category, and Product
sp = sales.groupby(["Product Category", "Sub Category", "Product"]).size().reset_index().rename(columns={0: "Count"})
sp.sort_values(by="Count", ascending=False)
```

```python
# Pivot table to see counts of products in each sub-category within product categories
sales.pivot_table(index=["Product Category"], columns=["Sub Category"], aggfunc="size").fillna(0)
```

```python
# Calculate total profit by Product Category and Sub Category
sales.groupby(["Product Category", "Sub Category"])['Profit'].sum().reset_index().sort_values(by="Profit", ascending=False) 
```
