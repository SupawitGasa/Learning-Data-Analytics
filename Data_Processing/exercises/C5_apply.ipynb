## Primer


```python
df = pd.DataFrame(
    {
        "Name": ["John", "Rock", "Neil", "Tom"],
        "Group": [
            "A",
            "A",
            "B",
            "B",
        ],
        "Marks": [78, 66, 55, 42],
    }
)
display(df)
```

```python
# Define a function to convert marks to percentage string
def make_percent(val):
    return str(val)+"%"


# Apply the function to create a new column "Marks (%)"
df["Marks (%)"] = df["Marks"].apply(make_percent)
```

```python
display(df)
```

```python
# Create a new column "Pass" which contains True if the student has passed the exam, False otherwise
# A student is considered to be passed if he/she has scored 70 or more marks in group A and 50 or more marks in group B.

# Define a function to determine if a student has passed
def is_pass(row):
    
    group = row["Group"]
    marks = row["Marks"]

    if group == "A":
        if marks >= 70:
            return True
        else:
            return False
    else:
        if marks >= 50:
            return True
        else:
            return False

# Apply the function to create the "Pass" column
df["Pass"] = df.apply(is_pass, axis=1)

display(df)
```

```python
# Load the student performance dataset
df_raw = pd.read_csv("./data/student_mat.csv")

# Display the info of the raw dataframe
df_raw.head()
```

```python
# For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column
df_stu = df_raw.loc[:, "school":"guardian"]
df_stu
```

```python
# Create a function that will capitalize strings
def capitalizer(val):
    return str(val).upper()


# Apply the function to the 'Mjob' and 'Fjob' columns
df_stu["Mjob"] = df_stu["Mjob"].apply(capitalizer)
df_stu["Fjob"] = df_stu["Fjob"].apply(capitalizer)
```

```python
display(df_stu)
```

```python
# Create a function that will return True if the age is greater than 17, else False
def is_legal_drinker(val):
    return val > 17


# Apply the function to create the "legal_drinker" column
df_stu["legal_drinker"] = df_stu["age"].apply(is_legal_drinker)

display(df_stu)
```

```python
# Create a function that categorizes students based on their age and sex
# If age < 18 and sex = 'F' => 'Underage Female'
# If age < 18 and sex = 'M' => 'Underage Male'
# If age >= 18 and sex = 'F' => 'Adult Female'
# If age >= 18 and sex = 'M' => 'Adult Male'
def categorize_by_age_and_sex(row):
    age = row["age"]
    sex = row["sex"]

    if (age < 18) and (sex == 'F'):
        return "Underage Female"
    elif (age < 18) and (sex == 'M'):
        return "Underage Male"
    elif (age > 18) and (sex == 'F'):
        return "Adult Female"
    else:
        return "Adult Male"


# Apply the function to create the "category" column
df_stu["category"] = df_stu.apply(categorize_by_age_and_sex, axis=1)

# Display the relevant columns
display(df_stu[["sex", "age", "category"]])
```

### Apply with groupby


```python
# Group the dataframe by the new column "category"
df_stu_grouped = df_stu.groupby(by="category")
```

```python
# Create a function that calculates the average age of students 
# whose father is a teacher vs those whose father is not a teacher
def average_age_of_teacher(df):
    filt_teacher = df["Fjob"] == "TEACHER"
    mean_age_teacher = df[filt_teacher]["age"].mean()
    mean_age_not_teacher = df[~filt_teacher]["age"].mean()
    return pd.Series(
        {
            "Mean Age Teacher": mean_age_teacher,
            "Mean Age Not Teacher": mean_age_not_teacher
        }
    )


# # # Testing the function on a single group
# group_list = list(df_stu_grouped.groups.keys())
# # print(group_list)

# df = df_stu_grouped.get_group(group_list[0])
# # display(df)

# filt_teacher = df["Fjob"] == "TEACHER"
# mean_age_teacher = df.loc[filt_teacher, "age"].mean()
# mean_age_not_techer = df.loc[~filt_teacher, "age"].mean()

# pd.Series(
#     {
#         "Mean Age Techer": mean_age_teacher, 
#         "Mean Age Not Teacher": mean_age_not_techer
#     }
# )

# Actually apply the function
df_stu_grouped.apply(average_age_of_teacher, include_groups=False)
```
